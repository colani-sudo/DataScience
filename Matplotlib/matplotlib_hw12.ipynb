```python
# Matplotlib/matplotlib_hw12.ipynb
# Created by Colani Matsenjwa on 2020/08/20 at 12:00

# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Reading the data from the csv file
data = pd.read_csv('data.csv')

data.head()
```

```python
# Creating data lists for Line chart of male, female, and total population in each township
township = []
male = []
female = []
population = []
for i in range(len(data['Township'])):
    township.append(data['Township'][i])
    male.append(data['Male'][i])
    female.append(data['Female'][i])
    population.append(data['Total(Male+Female)'][i])
print(township)
print(male)
print(female)
print(population)
```

```python
# Create subplots for the line chart
# plt.figure(figsize=(30, 30))        # Set the figure size
# plt.subplots_adjust(hspace=2, wspace=2)
```

```python
# Creating a bar chart for the population of the townships
plt.plot(township, population, color='green', linestyle='dashed', label='Population')
plt.plot(township, male, color='blue', linestyle='dashed', label='Male')
plt.plot(township, female, color='orange', linestyle='dashed', label='Female')

# Set the x-axis tick labels and rotate them diagonally
plt.xticks(ticks=range(len(township)), labels=township, rotation=60)

plt.title('the line graph of population')
plt.xlabel('Townships')
plt.ylabel('Number of people')

plt.legend()
plt.show()
```

```python
# Creating data lists for Line chart of immigration and emigration
immigration = []
emigration = []
for i in range(len(data['Township'])):
    immigration.append(data['Immigration'][i])
    emigration.append(data['Emigration'][i])
print(immigration)
print(emigration)
```

```python
# Creating Line chart for immigration and emigration data
plt.plot(township, immigration, color='blue', linestyle='solid', label='Immigration')
plt.plot(township, emigration, color='orange', linestyle='dashed', label='Emigration')

# Set the x-axis tick labels and rotate them diagonally
plt.xticks(ticks=range(len(township)), labels=township, rotation=60)

plt.title('the line graph of immigration and emigration')
plt.xlabel('Townships')
plt.ylabel('Number of people')

plt.legend()
plt.show()

```

```python
# Creating data lists for the Bar chart of household number
household = []
for i in range(len(data['Township'])):
    household.append(data['Number_of_Households'][i])

# Creating Bar chart for household number
plt.bar(township, household, color='green', label='Household')

# Set the x-axis tick labels and rotate them diagonally
plt.xticks(ticks=range(len(township)), labels=township, rotation=60)

plt.title('the long bar graph of household number')
plt.xlabel('Townships')
plt.ylabel('Number of people')

plt.legend()
plt.show()
```
