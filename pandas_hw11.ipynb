```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
df[::7]
```

```python
x = df[(df['High school system'] == 'National') & (df['Gender'] == 'male')].sum()
x
```

```python
keys = ['National', "county & city", "private"]
report = pd.DataFrame(columns=['High school system', 'Gender', 'Num of graduates', 'Num of enroll in police university',
                               'Num of enroll in police college', 'Military school enrollment average',
                               'Police university enrollment rate','Police collage enrollment rate', 'Military school enrollment rate'])
for k in range(len(keys)):
    key = keys[k]
    #the sums of national males and females
    sum_of_national_males = df[(df['High school system'] == key) & (df['Gender'] == 'male')].sum()
    sum_of_national_females = df[(df['High school system'] == key) & (df['Gender'] == 'female')].sum()

    #average of national males and females 
    average_of_national_males = round(df[(df['High school system'] == key) & (df['Gender'] == 'male')].mean(), 4)
    average_of_national_females = round(df[(df['High school system'] == key) & (df['Gender'] == 'female')].mean(), 4)

    # rate of national males and females
    #Police university rate
    police_uni_rate_nat_males = round(sum_of_national_males[5] / sum_of_national_males[2], 4)
    police_uni_rate_nat_females = round(sum_of_national_females[5] / sum_of_national_females[2], 4)

    #Police collage rate
    police_col_rate_nat_males = round(sum_of_national_males[6] / sum_of_national_males[2], 4)
    police_col_rate_nat_females = round(sum_of_national_females[6] / sum_of_national_females[2], 4)

    #Military school
    mili_sch_rate_nat_males = round(sum_of_national_males[7] / sum_of_national_males[2], 4)
    mili_sch_rate_nat_females = round(sum_of_national_females[7] / sum_of_national_females[2], 4)
    
    # Create a new row with the values
    new_row_male = {'High school system': key, 'Gender':"male", 'Num of graduates':sum_of_national_males[2],
                    'Num of enroll in police university':sum_of_national_males[5], 
                    'Num of enroll in police college': sum_of_national_males[6], 
                    'Military school enrollment average': average_of_national_males[-2],
                    'Police university enrollment rate':police_uni_rate_nat_males,
                    'Police collage enrollment rate':police_col_rate_nat_males, 
                    'Military school enrollment rate': mili_sch_rate_nat_males}
    
    new_row_female = {'High school system': key, 'Gender':"female", 'Num of graduates':sum_of_national_females[2],
                'Num of enroll in police university':sum_of_national_females[5], 
                'Num of enroll in police college': sum_of_national_females[6], 
                'Military school enrollment average': average_of_national_females[-2],
                'Police university enrollment rate':police_uni_rate_nat_females,
                'Police collage enrollment rate':police_col_rate_nat_females, 
                'Military school enrollment rate': mili_sch_rate_nat_females}

    # Insert the new row into the DataFrame
    report = report.append(new_row_male, ignore_index=True)
    report = report.append(new_row_female, ignore_index=True)
```
