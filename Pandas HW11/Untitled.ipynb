```python
import pandas as pd

df = pd.read_csv("data.csv")
output_df = df[["High school system", "Gender", "Num of graduates", "Num of enroll in police university", "Num of enroll in police college"]]
output_df["Military schoool enrollment average"] = (df["Num of enroll in military school"] / df["High school system"].value_counts().sum()).round(4)
output_df["Police university enrollment rate"] = (output_df["Num of enroll in police university"] / output_df["Num of graduates"]).round(4)
output_df["Police college enrollment rate"] = (df["Num of enroll in police college"] / df["Num of graduates"]).round(4)
output_df["Millitary school enrollment rate"] = (df["Num of enroll in military school"] / df["Num of graduates"]).round(4)

output_df.head()
```
